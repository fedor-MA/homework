```python
import numpy as np
import matplotlib.pyplot as plt

def beauty_derivative_plot(x0):
    """
       f(x) = 2sin(x) - cos(2x),  
         x0   [-2, 2]
    """

    def f(x):
        return 2 * np.sin(x) - np.cos(2 * x)
    
    def f_prime(x):
        return 2 * np.cos(x) + 2 * np.sin(2 * x)
    
    x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
    y = f(x)
    y_prime = f_prime(x)
    
    
    y0 = f(x0)
    y0_prime = f_prime(x0)
    
    #  
    tangent_line = y0_prime * (x - x0) + y0
    
    #  
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))
    
    #  :   
    ax1.plot(x, y, 'b-', linewidth=2, label=r'$f(x) = 2\sin x - \cos 2x$')
    ax1.plot(x, tangent_line, 'r--', linewidth=1.5, 
             label=f'  x = {x0:.2f}')
    ax1.plot(x0, y0, 'ro', markersize=8, label=' ')
    
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    ax1.set_title(f'     x = {x0:.2f}')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    ax1.axhline(y=0, color='k', linewidth=0.5)
    ax1.axvline(x=0, color='k', linewidth=0.5)
    
    #  : 
    ax2.plot(x, y_prime, 'g-', linewidth=2, 
             label=r"$f'(x) = 2\cos x + 2\sin 2x$")
    ax2.plot(x0, y0_prime, 'ro', markersize=8, 
             label=f'  x = {x0:.2f} = {y0_prime:.2f}')
    
    ax2.set_xlabel('x')
    ax2.set_ylabel("f'(x)")
    ax2.set_title(' ')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    ax2.axhline(y=0, color='k', linewidth=0.5)
    ax2.axvline(x=0, color='k', linewidth=0.5)
    
    plt.tight_layout()
    plt.show()

beauty_derivative_plot(-3)
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("C:\\Users\\ASUS\\Documents\\Prac\\homework_1\\populations.txt")
year = data[:, 0].astype(int)  # 
hares = data[:, 1]              # 
lynxes = data[:, 2]             # 
carrots = data[:, 3]             # 

#     
plt.figure(figsize=(15, 10))

plt.plot(year, hares, 'o-', label='Hares')
plt.plot(year, lynxes, 's-', label='Lynxes')
plt.plot(year, carrots, '^-', label='Carrots')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Species populations over time')
plt.legend()
plt.grid(True, alpha=0.3)

#    
max_hare_year = year[np.argmax(hares)]
max_lynx_year = year[np.argmax(lynxes)]
max_carrot_year = year[np.argmax(carrots)]

plt.axvline(max_hare_year, color='blue', alpha=0.3, linestyle='--')
plt.axvline(max_lynx_year, color='orange', alpha=0.3, linestyle='--')
plt.axvline(max_carrot_year, color='green', alpha=0.3, linestyle='--')
```

```python
total_hl = hares + lynxes
hares_ratio = hares / total_hl * 100
lynxes_ratio = lynxes / total_hl * 100

plt.bar(year, hares_ratio, label='Hares %', alpha=0.8)
plt.bar(year, lynxes_ratio, bottom=hares_ratio, label='Lynxes %', alpha=0.8)
plt.xlabel('Year')
plt.ylabel('Percentage (%)')
plt.title('Hares vs Lynxes ratio (100% stacked)')
plt.legend()
plt.grid(True, alpha=0.3, axis='y')
```

```python
# 
plt.subplot(3, 3, 1)
plt.hist(hares, bins=8, color='blue', alpha=0.7, edgecolor='black')
plt.axvline(np.mean(hares), color='red', linewidth=2, 
            label=f'Mean: {np.mean(hares):.0f}')
plt.axvline(np.mean(hares) + np.std(hares), color='orange', linestyle='--', 
            label=f'Std: {np.std(hares):.0f}')
plt.axvline(np.mean(hares) - np.std(hares), color='orange', linestyle='--')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title(f'Hares\n={np.mean(hares):.0f}, ={np.std(hares):.0f}')
plt.legend()

# 
plt.subplot(3, 3, 2)
plt.hist(lynxes, bins=8, color='orange', alpha=0.7, edgecolor='black')
plt.axvline(np.mean(lynxes), color='red', linewidth=2,
            label=f'Mean: {np.mean(lynxes):.0f}')
plt.axvline(np.mean(lynxes) + np.std(lynxes), color='orange', linestyle='--',
            label=f'Std: {np.std(lynxes):.0f}')
plt.axvline(np.mean(lynxes) - np.std(lynxes), color='orange', linestyle='--')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title(f'Lynxes\n={np.mean(lynxes):.0f}, ={np.std(lynxes):.0f}')
plt.legend()

# 
plt.subplot(3, 3, 3)
plt.hist(carrots, bins=8, color='green', alpha=0.7, edgecolor='black')
plt.axvline(np.mean(carrots), color='red', linewidth=2,
            label=f'Mean: {np.mean(carrots):.0f}')
plt.axvline(np.mean(carrots) + np.std(carrots), color='orange', linestyle='--',
            label=f'Std: {np.std(carrots):.0f}')
plt.axvline(np.mean(carrots) - np.std(carrots), color='orange', linestyle='--')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title(f'Carrots\n={np.mean(carrots):.0f}, ={np.std(carrots):.0f}')
plt.legend()
```

```python
winner = np.where(hares > lynxes, 'H', 'L')
print("  (H = hares, L = lynxes):", winner.tolist())

#  
winner_counts = np.array([np.sum(winner == 'H'), np.sum(winner == 'L')])
plt.pie(winner_counts, labels=['Hares win', 'Lynxes win'], autopct='%1.1f%%',
        colors=['blue', 'orange'], startangle=90)
plt.title('Who has larger population each year?')
```

```python
hares_change = np.diff(hares)
lynxes_change = np.diff(lynxes)
years_change = year[1:]  #      

plt.plot(years_change, hares_change, 'o-', label='Hares change')
plt.plot(years_change, lynxes_change, 's-', label='Lynxes change')
plt.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
plt.xlabel('Year')
plt.ylabel('Population change')
plt.title(f'Yearly changes\nCorrelation: {np.corrcoef(hares_change, lynxes_change)[0,1]:.3f}')
plt.legend()
plt.grid(True, alpha=0.3)
```

```python
plt.scatter(hares, lynxes, c=year, cmap='viridis', s=100, alpha=0.8)
plt.colorbar(label='Year')
plt.xlabel('Hares population')
plt.ylabel('Lynxes population')
plt.title('Hares vs Lynxes by year')
plt.grid(True, alpha=0.3)

#     
z = np.polyfit(hares, lynxes, 1)
p = np.poly1d(z)
plt.plot(hares, p(hares), "r--", alpha=0.8, label=f'Corr: {np.corrcoef(hares, lynxes)[0,1]:.3f}')
plt.legend()

plt.tight_layout()
plt.show()
```
