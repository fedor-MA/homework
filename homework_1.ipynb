```python
def rotate_matrix(matrix, direction):
    if direction == 'right':
        return [list(row)[::-1] for row in zip(*matrix)]
    elif direction == 'left':
        return [list(row) for row in zip(*matrix)][::-1]
    else:
        raise ValueError("Направление должно быть 'left' или 'right'")
    
    
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("Исходная матрица:")
for row in matrix:
    print(row)

print("\nПоворот вправо (по часовой стрелке):")
for row in rotate_matrix(matrix, 'right'):
    print(row)

print("\nПоворот влево (против часовой стрелки):")
for row in rotate_matrix(matrix, 'left'):
    print(row)
```

```python
from datetime import datetime
from typing import List, Optional, Dict, Any

class Book:
    
    def __init__(self, title: str, author: str, year: int, isbn: str, copies: int):
        
        self.title = title
        self.author = author
        self.year = year
        self.isbn = isbn
        self.copies = copies
    
    def get_info(self):
        
        return {
            'title': self.title,
            'author': self.author,
            'year': self.year,
            'isbn': self.isbn,
            'copies': self.copies
        }
    
    def change_copies(self, delta: int):

        if self.copies + delta >= 0:
            self.copies += delta
            return True
        return False
    
    def __str__(self):
        return f"'{self.title}' {self.author} ({self.year}) - {self.copies} экз."


class User:
    
    def __init__(self, name: str, card_number: str):
       
        self.name = name
        self.card_number = card_number
        self.borrowed_books: List[Book] = []
    
    def register(self):

        return {
            'name': self.name,
            'card_number': self.card_number,
            'borrowed_count': len(self.borrowed_books)
        }
    
    def add_book(self, book: Book):

        if book not in self.borrowed_books:
            self.borrowed_books.append(book)
            return True
        return False
    
    def remove_book(self, book: Book):
        
        if book in self.borrowed_books:
            self.borrowed_books.remove(book)
            return True
        return False
    
    def __str__(self):
        
        books_str = ', '.join([book.title for book in self.borrowed_books]) or 'нет книг'
        return f"{self.name} (билет №{self.card_number}): {books_str}"


class Library:

    def __init__(self, name: str):

        self.name = name
        self.books: List[Book] = []
        self.users: List[User] = []
        self.transaction_history: List[Dict[str, Any]] = []
    
    def add_book(self, book: Book):
        
        self.books.append(book)
        print(f"Книга '{book.title}' добавлена в библиотеку")
    
    def remove_book(self, isbn: str):

        for book in self.books:
            if book.isbn == isbn:
                self.books.remove(book)
                print(f"Книга '{book.title}' удалена из библиотеки")
                return True
        print(f"Книга с ISBN {isbn} не найдена")
        return False
    
    def add_user(self, user: User):

        self.users.append(user)
        print(f"Пользователь '{user.name}' зарегистрирован")
    
    def remove_user(self, card_number: str):

        for user in self.users:
            if user.card_number == card_number:
                if user.borrowed_books:
                    print(f"Невозможно удалить: пользователь '{user.name}' имеет невозвращенные книги")
                    return False
                self.users.remove(user)
                print(f"Пользователь '{user.name}' удален")
                return True
        print(f"Пользователь с билетом №{card_number} не найден")
        return False
    
    def find_books(self, **kwargs):

        results = self.books.copy()
        
        for key, value in kwargs.items():
            if key == 'title':
                results = [b for b in results if value.lower() in b.title.lower()]
            elif key == 'author':
                results = [b for b in results if value.lower() in b.author.lower()]
            elif key == 'year':
                results = [b for b in results if b.year == value]
            elif key == 'isbn':
                results = [b for b in results if b.isbn == value]
        
        return results
    
    def find_users(self, **kwargs):

        results = self.users.copy()
        
        for key, value in kwargs.items():
            if key == 'name':
                results = [u for u in results if value.lower() in u.name.lower()]
            elif key == 'card_number':
                results = [u for u in results if u.card_number == value]
        
        return results
    
    def issue_book(self, card_number: str, isbn: str):
    
        users = self.find_users(card_number=card_number)
        if not users:
            print(f"Пользователь с билетом №{card_number} не найден")
            return False
        user = users[0]
    
        books = self.find_books(isbn=isbn)
        if not books:
            print(f"Книга с ISBN {isbn} не найдена")
            return False
        book = books[0]
        
        if book.copies <= 0:
            print(f"Нет доступных экземпляров книги '{book.title}'")
            return False
        
        if book.change_copies(-1) and user.add_book(book):
            transaction = {
                'type': 'ISSUE',
                'user': user.name,
                'book': book.title,
                'date': datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                'details': f"Книга выдана пользователю {user.name}"
            }
            self.transaction_history.append(transaction)
            print(f"Книга '{book.title}' выдана пользователю {user.name}")
            return True
        
        return False
    
    def return_book(self, card_number: str, isbn: str):
       
        users = self.find_users(card_number=card_number)
        if not users:
            print(f"Пользователь с билетом №{card_number} не найден")
            return False
        user = users[0]
        
        
        books = self.find_books(isbn=isbn)
        if not books:
            print(f"Книга с ISBN {isbn} не найдена")
            return False
        book = books[0]
        
        if book not in user.borrowed_books:
            print(f"У пользователя {user.name} нет книги '{book.title}'")
            return False
        
        if book.change_copies(1) and user.remove_book(book):
            transaction = {
                'type': 'RETURN',
                'user': user.name,
                'book': book.title,
                'date': datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                'details': f"Книга возвращена пользователем {user.name}"
            }
            self.transaction_history.append(transaction)
            print(f"Книга '{book.title}' возвращена пользователем {user.name}")
            return True
        
        return False
    
    def show_transaction_history(self) -> None:

        print("\n" + "="*60)
        print(f"ИСТОРИЯ ОПЕРАЦИЙ БИБЛИОТЕКИ '{self.name}'")
        print("="*60)
        
        if not self.transaction_history:
            print("История операций пуста")
            return
        
        for i, transaction in enumerate(self.transaction_history, 1):
            print(f"{i}. [{transaction['date']}] {transaction['type']}: {transaction['details']}")
        
        print("="*60)
    
    def show_all_books(self):

        print(f"\nКНИГИ В БИБЛИОТЕКЕ '{self.name}':")
        if not self.books:
            print("  В библиотеке нет книг")
        else:
            for i, book in enumerate(self.books, 1):
                print(f"  {i}. {book}")
    
    def show_all_users(self):
        
        print(f"\nПОЛЬЗОВАТЕЛИ БИБЛИОТЕКИ '{self.name}':")
        if not self.users:
            print("  В библиотеке нет пользователей")
        else:
            for i, user in enumerate(self.users, 1):
                print(f"  {i}. {user}")


def main():
    print("СОЗДАНИЕ БИБЛИОТЕКИ")
    print("-" * 40)
    
    
    library = Library("Библиотека №1")
    print(f"Библиотека '{library.name}' создана")
    
    print("\nДОБАВЛЕНИЕ КНИГ")
    print("-" * 40)
    
    
    book1 = Book("Война и мир", "Лев Толстой", 1869, "978-5-699-38291-8", 5)
    
   
    library.add_book(book1)
    
    print("\nДОБАВЛЕНИЕ ПОЛЬЗОВАТЕЛЕЙ")
    print("-" * 40)
    
   
    user1 = User("Иван Петров", "LIB-001")

    
    
    library.add_user(user1)
    
    
    print("\nТЕКУЩЕЕ СОСТОЯНИЕ БИБЛИОТЕКИ")
    print("-" * 40)
    library.show_all_books()
    library.show_all_users()
    
    print("\nСЦЕНАРИЙ ВЫДАЧИ КНИГ")
    print("-" * 40)
    
   
    library.issue_book("LIB-001", "978-5-699-38291-8")
     
    
    print("\nПОПЫТКА ВЗЯТЬ НЕСУЩЕСТВУЮЩУЮ КНИГУ")
    print("-" * 40)
    library.issue_book("LIB-002", "111-111-111")
    
    print("\nПОПЫТКА ВЗЯТЬ КНИГУ, КОТОРОЙ НЕТ В НАЛИЧИИ")
    print("-" * 40)
    
    rare_book = Book("Редкая книга", "Неизвестный автор", 1900, "978-5-00000-000-0", 0)
    library.add_book(rare_book)
    library.issue_book("LIB-003", "978-5-00000-000-0") 
    print("\nСОСТОЯНИЕ ПОСЛЕ ВЫДАЧИ КНИГ")
    print("-" * 40)
    library.show_all_books()
    library.show_all_users()
    
    print("\nПОИСК КНИГ")
    print("-" * 40)
    
  
    tolstoy_books = library.find_books(author="Толстой")
    print("Книги Льва Толстого:")
    for book in tolstoy_books:
        print(f"  - {book}")
    
    
    search_results = library.find_books(title="преступление")
    print("\nКниги с 'преступление' в названии:")
    for book in search_results:
        print(f"  - {book}")
    
    print("\nСЦЕНАРИЙ ВОЗВРАТА КНИГ")
    print("-" * 40)
    

    library.return_book("LIB-001", "978-5-699-38291-8")
    
    print("\nПОПЫТКА ВЕРНУТЬ КНИГУ, КОТОРУЮ ПОЛЬЗОВАТЕЛЬ НЕ БРАЛ")
    print("-" * 40)
    library.return_book("LIB-002", "978-5-699-38291-8")
    
    print("\nСОСТОЯНИЕ ПОСЛЕ ВОЗВРАТА КНИГ")
    print("-" * 40)
    library.show_all_books()
    library.show_all_users()
    
    print("\nПОПЫТКА УДАЛИТЬ ПОЛЬЗОВАТЕЛЯ С НЕВОЗВРАЩЕННЫМИ КНИГАМИ")
    print("-" * 40)
    library.remove_user("LIB-001")
    
    print("\nУДАЛЕНИЕ ПОЛЬЗОВАТЕЛЯ БЕЗ ДОЛГОВ")
    print("-" * 40)
    library.return_book("LIB-001", "978-5-17-112799-7")
    library.remove_user("LIB-001")
    library.show_transaction_history()
    
    print("\nФИНАЛЬНОЕ СОСТОЯНИЕ БИБЛИОТЕКИ")
    print("-" * 40)
    library.show_all_books()
    library.show_all_users()


if __name__ == "__main__":
    main()
```
